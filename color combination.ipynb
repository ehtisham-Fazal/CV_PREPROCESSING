```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
image = cv2.imread(r'C:\Users\ehtis\OneDrive\Desktop\extra\image.jpg')
```

```python
def imshow(title="image", image=None, size=10):
    w,h=image.shape[0], image.shape[1]
    aspect_ratio=w/h
    plt.figure(figsize=(aspect_ratio*size,size))
    plt.imshow(cv2.cvtColor(image,cv2.COLOR_RGB2BGR))
    plt.title(title)
    plt.show()
```

```python
B,G,R=cv2.split(image)
```

```python
print(B.shape,G.shape,R.shape)
```

```python
imshow("only red",R)
```

```python
zeros=np.zeros(image.shape[:2], dtype="uint8")
print(zeros.shape)
# imshow("red",image[zeros,zeros,R])
```

```python
imshow("red",cv2.merge([zeros,zeros,R]))
```

```python
imshow("Green",cv2.merge([zeros,G,zeros]))
```

```python
imshow("blue",cv2.merge([B,zeros,zeros]))
```

```python
imshow("blue",cv2.merge([B+100,G,R+10]))
```
